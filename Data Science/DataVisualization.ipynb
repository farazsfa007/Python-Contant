```python
import matplotlib.pyplot as plt
```

```python
df.loc[['India','Philippines','Viet Nam','China'],Years].T.plot( figsize=(20,7) )
```

```python
df.loc[['India','Philippines','Viet Nam','China'],Years].T.plot( figsize=(20,7) )
plt.title('Immegration Timeline Of Asian Country',size=20)
plt.xlabel('Years',size=15)
plt.ylabel('No. of Immegrants',size=15)
plt.xticks(rotation=40)
```

## BAR CHART

```python
top_5_europe=df[df['AreaName']=='Europe'].sort_values('1990',ascending=False)['1990'].head()
```

```python
top_5_europe.plot(kind='bar',figsize=(20,8))
plt.title('Top 5 Contries Of Europe',size=20)
plt.xlabel('Country name',size=15)
plt.ylabel('Immegration',size=15)
plt.xticks(rotation=360)
```

```python
df['Total']=df.sum(axis=1)
```

```python
continent_data=df.groupby('AreaName').sum()['Total']
```

```python
continent_data.plot(kind='pie',autopct='%1.2f%%',figsize=(20,10))
```

```python
continent_data.plot(kind='pie',autopct='%1.2f%%',figsize=(20,10),labels=None)
plt.legend(continent_data.index)
```
